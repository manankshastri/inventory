```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
import requests
import pandas as pd
import re
```

```python
data = pd.read_csv("airtableFinal.csv")
```

```python
def urlR(index, front=False):
  if front:
    fileName = data.at[index, 'Front Screenshots'].split()[-1]
  else:
    fileName = data.at[index, 'Back Screenshots'].split()[-1]
  url = re.sub("['(', ')']", "", fileName)
  image = Image.open(requests.get(url, stream=True).raw)
  image = np.asarray(image)
  image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
  # plt.imshow(image)
  # plt.show()
  return image
```

```python
from skimage.metrics import structural_similarity as compare_ssim
```

```python
def checkNull(index):
  if not pd.isnull(data.at[index, 'Front Screenshots']) and not pd.isnull(data.at[index, 'Back Screenshots']):
    return True
  else:
    return False
```

```python
def checkSize(image):
  image1 = cv2.copyMakeBorder(image.copy(), 30, 30, 30, 30, cv2.BORDER_CONSTANT, value=0)
  ret, binary = cv2.threshold(image1, 50, 200, cv2.THRESH_BINARY_INV)

  contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
  sortedCont = sorted(contours, key=cv2.contourArea)

  return cv2.contourArea(sortedCont[-2])
```

```python
same = []
```

```python
def pop(index):
  frontImage = urlR(index, True)
  backImage = urlR(index, False)
  
  frontImageSize = checkSize(frontImage)
  backImageSize = checkSize(backImage)

  if frontImageSize == backImageSize:
    (score, diff) = compare_ssim(frontImage, backImage, full=True)
    diff = (diff*255).astype('uint8')
    return score
```
